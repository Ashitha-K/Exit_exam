Preprocessing

```python
#Preprocessing
#Fillig missing values
data.isnull().sum()
```

```python
#Filling the numerical values with median and categorical values with mode
for i in data.columns:
  if data[i].dtype=='object':
    data[i]=data[i].fillna(data[i].mode()[0])
  else:
    data[i]=data[i].fillna(data[i].median())
```

```python
data.isnull().sum()
```

```python
#Finding out the outliers for numerical columns
for i in data.columns:
  if data[i].dtype=='int64' or data[i].dtype=='float64':
    sns.boxplot(data[i])
    plt.show()
```

```python
outlier_checking = ['Price','Bathroom','Bedroom','Land','Longitude','Latitude','Building']
Q1 = data[outlier_checking].quantile(0.25)
Q3 = data[outlier_checking].quantile(0.75)
IQR = Q3 - Q1
outliers_lower = data[outlier_checking]<(Q1-1.5*IQR)
outliers_upper = data[outlier_checking]>(Q3 + 1.5*IQR)
print(data[outlier_checking])
```

```python
#treating the outlies using IQR method
data[outlier_checking] = data[outlier_checking].clip(lower = Q1-1.5*IQR, upper = Q3+1.5*IQR, axis = 1)
```

```python
#label encoding
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
data_encoded=data.copy()
data_encoded['City/Regency']=le.fit_transform(data_encoded['City/Regency'])
```

```python
data_encoded.head()
```

```python

#Choosing the location with high no.of counts and grouping the other location to location_others
# Correct use of lambda function to encode specific locations
data_encoded['Location_Buah Batu'] = data_encoded['Location'].apply(lambda x: 'Buah Batu' if x == 'Buah Batu' else 'Location_others')
data_encoded['Location_Padalarang'] = data_encoded['Location'].apply(lambda x: 'Padalarang' if x == 'Padalarang' else 'Location_others')
data_encoded['Location_Sukasari'] = data_encoded['Location'].apply(lambda x: 'Sukasari' if x == 'Sukasari' else 'Location_others')
data_encoded['Location_Margaasih'] = data_encoded['Location'].apply(lambda x: 'Margaasih' if x == 'Margaasih' else 'Location_others')

```

```python
#Defining x and y
x=data_encoded.drop(['Price'],axis=1)
y=data_encoded['Price']
```

```python
data_encoded=pd.get_dummies(data_encoded)
```

```python
#deleting the location
data_encoded.drop('Location',axis=1,inplace=True)
```

```python
#splitting
x=data_encoded.drop(['Price'],axis=1)
y=data_encoded['Price']
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#importing linear regression
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
#importing the metrics
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
y_pred=lr.predict(x_test)
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
```
